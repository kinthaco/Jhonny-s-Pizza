```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import StaleElementReferenceException, TimeoutException

# Specify the path to the Chrome user profile directory (parent directory of 'Default')
chrome_profile_path = r'C:\Users\chak\AppData\Local\Google\Chrome\User Data'  # Updated path
download_dir = r'C:\Users\chak\OneDrive\Desktop\BIA Term-4\Capstone\Dataset'

# Set up Chrome options to use the existing profile
chrome_options = Options()
chrome_options.add_argument(f"user-data-dir={chrome_profile_path}")  # Set user data directory
chrome_options.add_argument("profile-directory=Default")  # Use the default profile

#Setting the path to download files
path_ref = {'download.default_directory': download_dir}
chrome_options.add_experimental_option("prefs", path_ref)

# Path to ChromeDriver 
chrome_driver_path = r'C:\Users\msgok\Downloads\chromedriver-win64\chromedriver-win64\chromedriver.exe'  # Your actual path

# Initialize the WebDriver with the profile options
service = Service(chrome_driver_path)
driver = webdriver.Chrome(service=service, options=chrome_options)
```

```python
# Links of webpage that needs to be accessed
links = ['https://app.squareup.com/dashboard/sales/reports/sales-summary', 
         'https://app.squareup.com/dashboard/sales/reports/payment-methods', 
         {'https://app.squareup.com/dashboard/sales/reports/item-sales':["//market-button-dropdown[@class='market-button-dropdown']", 
                                                                        "//market-row[@class='market-row preload']", 
                                                                        "//market-row[@class='market-row preload hide-bottom-border']"]}, 
         {'https://app.squareup.com/dashboard/sales/reports/category-sales':"//market-button[@aria-label='Start export']"},
         'https://app.squareup.com/dashboard/sales/reports/section-sales', 
         {'https://app.squareup.com/dashboard/sales/reports/team-sales':"//market-row[@class='market-row preload']"}, 
         'https://app.squareup.com/dashboard/sales/reports/labor', 
         {'https://app.squareup.com/dashboard/sales/reports/modifiers':"//market-button[@aria-label='Start export']"},
         {'https://app.squareup.com/dashboard/sales/reports/voids':"//market-button[@aria-label='Start export']"}]
print(links)
```

```python
def wait_and_click(driver, by_method, locator, retries=3, timeout=20): # Function to explicitly wait and click a button
    for _ in range(retries): # No of retries when an exception occurs
        try:
            WebDriverWait(driver, timeout).until(EC.element_to_be_clickable((by_method, locator))).click() # explicit wait before clicking a button
            return 
        except StaleElementReferenceException as s: # Handling StaleElementReferenceException 
            print(f"The StaleElement error occured trying again")
        except TimeoutException as t: #Handling TimeoutException
            print("Timout Exception occured. Retrying...")
    raise Exception(f"Failed to click the button after {retries} times") # If an exception occurs for consecutively for 3 times program breaks


for i in range(275):
    for link_entry in links:
        # If it's a dictionary (link with xpaths), handle the XPaths
        if isinstance(link_entry, dict):
            link, xpath_values = list(link_entry.items())[0]
            driver.get(link) # Loading the webpage of interest
            
            # If there are multiple XPaths to click
            if isinstance(xpath_values, list):
                for xpath_value in xpath_values:
                    wait_and_click(driver, By.XPATH, xpath_value)
            else:  # If there is only a single XPath to click
                wait_and_click(driver, By.XPATH, "//market-button[@aria-label='Export']")
                wait_and_click(driver, By.XPATH, xpath_values)
        else:
            # Loading the webpage of interest
            driver.get(link_entry)
            wait_and_click(driver, By.XPATH, "//market-button[@aria-label='Export']")
    
    # Clicking the button to go to previous date
    wait_and_click(driver, By.ID, 'ember79')
```
