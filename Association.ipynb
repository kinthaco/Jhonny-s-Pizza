```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
dataset= pd.read_csv("Item_Transactions_apriori.csv")
dataset
```

```python
dataset.columns
```

```python
dataset.drop(columns='Unnamed: 0',inplace=True)
```

```python
transcation_dataset=dataset.drop(columns='Transaction_ID')
```

```python
transcation_dataset.columns
```

<h1> Applying Apriori

```python
transcation_dataset['Sausage and Peppers'].sum()
```

```python
transcation_dataset.shape
```

```python
contains_value=(dataset==3).any()
contains_value
```

```python
check_df=pd.DataFrame(columns=['Min_suppport','min_confidence','rules_greater#','rules_less#','highest_lift'])
check_df
```

```python
min_sup=[0.01,0.05,0.1]
conf=[0.45,0.48,0.5,0.51,0.52,0.53,0.54,0.55]
```

```python
check_df=check_df[0:0]
for value in min_sup:
    for no in conf:
        itemsets=apriori(transcation_dataset,min_support=value,use_colnames=True)
        rules=association_rules(itemsets,metric='confidence',min_threshold=no)
        check_df.loc[len(check_df)]=[value,no,len(rules[rules.lift>1]),len(rules[rules.lift<1]),max(rules.lift)]
check_df     
```

```python
frequent_itemset=apriori(transcation_dataset,min_support=0.05,use_colnames=True)
frequent_itemset
```

```python
rules=association_rules(frequent_itemset,metric="confidence",min_threshold=0.45)
rules
```

```python
rules.drop(['antecedent support',
 'consequent support','conviction',
 'zhangs_metric'],axis=1,inplace=True)
rules.sort_values(by=['lift'],ascending=False)
```
